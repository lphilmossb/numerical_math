```python
import numpy as np
from fft import fft
from scipy.fft import fftshift
import matplotlib.pyplot as plt
from trig_inter import trig_inter

t = np.array([0.25 * j for j in range(8)])
f1 = lambda t : t**4 -3*t**3 + 2*t**2 - np.tan(t**2 - 2*t)
X = fft(f1(t))
x = np.linspace(0, 2, 500)

interpolant = trig_inter(f1(t), x, dt=t[1] - t[0])

plt.plot(x,f1(x), label='f')
plt.plot(x, interpolant, label='p')
plt.legend()
plt.show()

# np.savetxt('ex_7_3_2.csv', np.column_stack((x, f1(x), interpolant)), header='x,f,p', comments='', delimiter=',')
```
