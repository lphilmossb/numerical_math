```python
import numpy as np
import matplotlib.pyplot as plt

from Fitter import Fitter
from pandas import DataFrame

plt.rcParams.update({
    'text.usetex': True,
    'grid.linestyle': '--',
    'font.size': 18,
})

x_data = np.array([-0.04, 0.93, 1.95, 2.90, 3.83, 5.00, 5.98, 7.05, 8.21, 9.08, 10.09])
y_data = np.array([-8.66, -6.44, -4.36, -3.27, -0.88, 0.87, 3.31, 4.63, 6.19, 7.40, 8.85])

fitter1 = Fitter(x_data, 1)
fitter2 = Fitter(x_data, 2)
fitter3 = Fitter(x_data, 3)

fitter1.PolyFit(y_data)
fitter2.PolyFit(y_data)
fitter3.PolyFit(y_data)

sigma1 = fitter1.StdDev(y_data)
sigma2 = fitter2.StdDev(y_data)
sigma3 = fitter3.StdDev(y_data)

fig, ax = plt.subplots()
fig.set_size_inches(12,12)
ax.grid()
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')

l1 = fitter1.PlotPoly(ax, y_data)[0]
l2 = fitter2.PlotPoly(ax)
l3 = fitter3.PlotPoly(ax)

l1.set_label(rf'$m=1$, $\sigma=${fitter1.StdDev(y_data):.4f}')
l2.set_label(rf'$m=2$, $\sigma=${fitter2.StdDev(y_data):.4f}')
l3.set_label(rf'$m=3$, $\sigma=${fitter3.StdDev(y_data):.4f}')
fig.suptitle('Plots of fitting polynomials for $m=1,2,3$')
fig.tight_layout()
ax.legend()
plt.show()

# fig.savefig('ex_5_3_iii.png', dpi=300)
```
