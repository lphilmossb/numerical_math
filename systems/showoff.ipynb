```python
from numpy import linspace, pi, arange, sin
import seaborn as sns
import matplotlib.pyplot as plt

from plot_util import *

mpl_setscheme()
fig, ax = get_figure(figsize=(12,10))

cmap = sns.color_palette('hls', as_cmap=True)
t = linspace(0, 6*pi, 1000)

def time2colour(t):
    return cmap((t % (2*pi)) / (2 * pi))

cols = time2colour(t)
y = sin(t)
xticks = arange(t[0], t[-1], pi * 0.5)
ax.set_xticks(xticks)
lc, fill = fillplot(ax, t, y, 0, cols, 'white')
lc.set_linewidth(3)
#fig.savefig('showoff.png')
plt.show()
```
