```python
from numpy import *
from ForwardEuler import ForwardEuler1D
from pandas import DataFrame
from plot_util import *

mpl_setscheme()


def dy(y, t):
    return 1.0 / ((1+t)**2) - y / (1+t)

def f(t):
    return (log(t + 1) + 1) / (t + 1)

h = 0.2
fwe = ForwardEuler1D(dy, 1, 0, 1 + h, h)


y, t = fwe.run()
errors = y - f(t)

data = DataFrame({'Forward Euler': y, 'Exact': f(t), 'Error' : errors})

display(data)

fig, ax = get_figure()
tfine = linspace(0,1,500)
ax.plot(t,y)
ax.plot(tfine, f(tfine))
```
